```python
# функций для преобразования изображения обратно в пристойный вид
def deprocess_img(processed_img):
  x = processed_img.copy()
  if len(x.shape) == 4:
    x = np.squeeze(x,0) # удалить нулевую ось
    assert len(x.shape) == 3, ('Image inputed to deprocess must have dimentions: ',
      '[1,height,width,channels] or [height,width,cchannels]')

    if len(x.shape) != 3:
      raise ValueError('Invalid input to image deprocessing')

    # отменить усреднение показателей RGB
    x[:,:,0] += 103.939
    x[:,:,1] += 116.779
    x[:,:,2] += 123.68

    # преобразование BGR to RGB
    x = x[:,:,::-1]

    # ограничить наибольшее и наименьшее значение у параметров RGB (0-255)
    x = np.clip(x,0,255).astype('uint8')

    return x
```

# Принцип стилизации изображений
В простейшем случае изображение стилизуется попиксельно. Т.е. как настраиваемые параметры алгоритма градиентного спуска мы берем числовые значения каждого пикселя. А как желамые параметры мы берем значения пикселей изображения, стиль которого берем. Однако это грубый метод, стилизует он плохо, и даже не отребует использования нейросетей.

Однако без нейросети НЕВОЗМОЖНО понять СТЕПЕНЬ и КОРЕКТНОСТЬ стилизации. Иными словами нейросеть используется лишь для вычисления критерия качества. В данной задаче лучшие сети - сверточные.

**Леон Гатес** в работе по исследованию этого вопроса разделил общий критерий качества на 2 независимые оценки (вычисляются они независимо):


1.   Степень соответствия преобразованного изображения исходному (Jc);
2.   Степень стилизации преобразованного изображения (Js).

Тогда общая формула вычисления критерия качества - **J = a\*Jc + b\*Js**
Где мнодители a и b характеризуют важность каждого из слагаемых.

# Расчет степени соответствия преобразованного изображения исходному (Jc) (Или расхождение по контенту)
Сверточная сеть (CNN) состоит из 2 частей, части со слоями свертки и части с полносвязной сетью. Выяснено, что если взять уже обученную сеть CNN, то её последний сверточный слой будет ничем иным как описанием общий признаков поданного изображения. Не сам по себе, а как результирующий тензор, полученный после подачи некого изображения на взод сети. Зная это можно не сравнивать исходное и стилизованное изображение попиксельно, а сравнить эти тензоры, тем самым сравнив общие описания этих изображений.

В сети VGG19 такой последний сверточный слой называется - Conv5-2. Его мы и возьмем для расчетов. Он имеет размерность 14\*14\*512. 14 - размеры карт признаков каждого канала, а 512 - количество каналов.

В таком случае чтобы вычислить рассогласование между исходным и преобразованным изображением мы пропустим их по очереди через сеть и получим два тензора T и P, где T описывает исходное изображение, а P преобразованное. Расчет их разности по большому счету просто поэлементное вычитание элементов тензоров, потом возведение каждого элемента в квадрат, потом их сложение, потом усреднение. Иными словами показатель Jc вычисляется по следующим шагам:

1. Матричное вычитание P - T
2. Каждый элемент полученной матрицы возводим в квадрат
3. Складываем все элементы полученной матрицы
4. И делим полученное число на произведение всех измерений матрицы

В нашем случае все матрицы будут иметь размерность 14\*14\*512.

# Расчет степени стилизации преобразованного изображения (Js)
В отличии от контента стиль - куда более эфимерное понятие, поэтому и расчитывается иначе. Считается, что изображение соответствует заданному стилю на многих уровнях, поэтому надо как то учесть это при расчетах. Как? - Путем учета в расчетах тензоров, полученных на выходах многих сверточных слоев, так как КАЖДЫЙ сверточный слой на выходе имеет тензор, описывающий поданное изображение. Разность между этими тензорами в уровне абстракции. Соответственно, чем дальше от входа модели находится сверточный слой, тем более абстрактным является его описание полученного изображения.

Любая сверточная сеть состоит из чередования сверточных слоев и слоев "пуллинга" (MaxPooling, MinPooling or AvgPooling). Однако свертчные слои не обязательно должны каждый раз чередоваться со слоями пуллинга. Можно поставить несколько сверточных слоев подряд. Зачем? - Эта операция позволит сымитировать анализ изображения большим ядром свертки, чем на самом деле и будем иметь меньшее количество настраиваемыз параметров, что ускорит обучение. Например применив подрад два сверточных слоя с ядром свертки 3\*3 мы сымитируем обработку изображения ядром 5\*5. При том, что количество настраиваемых параметров будет меньше чем у ядра 5\*5. 5\*5+1=26 против 2\*(3\*3+1)+1=20.

```python
# в vgg19 есть 19 слоев, каждый имеет свое имя
# ниже 2 коллекции имен, первая для слоев, по которым мы будем считать
# ошибку по контенту, второй - те, с помощью которыйх будем считать
# степень стилизации изображения
content_layers = ['block5_conv4']

style_layers = ['block1_conv1',
                'block2_conv1',
                'block3_conv1',
                'block4_conv1',
                'block5_conv1',
                ]

num_content_layers = len(content_layers)
num_style_layers = len(style_layers)                
```

```python
# функция для расчета потерь по контенту (Jc)
def get_content_loss(base_content, target):
  # base_content и target - тензоры размером 14*14*512
  # как и говорилось выше вычисляем ошибку по контенту
  # base_content - target - матричное вычитание тензоров
  # tf.square() - возведение в квадрат всех элементов тензоров
  # tf.reduce_mean() - нахождение среднего значения элемента в полученной матрице
  # т.е. сначала все элементы сложили, а потом поделили на их количество
  return tf.reduce_mean(tf.square(base_content - target))
```

```python
# функция для вычисляния матрицы Грамма для поданного тензора
def get_gram_matrix(input_tensor):
  # тензор на входе будет иметь размерность высота карты признаков * ширина карты признаков * 
  # количество каналов. Для расчета матрицы Грамма нам надо сначала "вытянуть этот тензор"
  # в двумерную матрицу, где количество столбцов - количество каналов поданного тензора
  # а количество строк - произведение двух других размерностей
  # сначала посчитаем кол-во каналов в поданном тензоре
  channels_num = int(input_tensor.shape[-1])

  # [-1, channels_num] - новая желаемая размерность тензора, т.е. матрица
  # написав -1 мы говорим, что нам не принципиально количество строк этой матрицы
  # но указав channels_num мы жестко устанавливаем количество её столбцов
  a = tf.reshape(input_tensor,[-1, channels_num])

  # получить кол-во строк в матрице
  rows_num = a.shape[0] 

  # умнажаем матрично вытянутый тензор на его транспонированную версию
  gram_matrix = tf.matmul(a,a, transpose_a=True)

  # tf.cast(rows_num, tf.float32) - просто число типа float32
  # делаая такую операцию мы просто усредняем все значения в матрице Грамма
  # ВАЖНО я не писал об этой необходимости ранее, плюс непонятно, почему надо делить
  # лишь на кол-во строк в вытянутой матрице
  return gram_matrix #/ tf.cast(rows_num, tf.float32)
```

```python
# функция для вычисления расхождения по стилю для заданного выходного слоя
def get_style_loss(current_style_tensor, target_gram_matrix):
  # current_style_tensor - тензор, полученный на выходе определенного слоя 
  # попсле обработки ФОРМИРУЕМОГО изображения
  # target_gram_matrix - матрица Грамма, посчитанная для изображения со стилем для данного слоя

  # посчитаем матрицу грамма для тензора формируемого изображения
  current_style_gram_matrix = get_gram_matrix(current_style_tensor)

  # как и описано выше поэллементно вычитаем две матрицы
  # возводим все элеементы полученной матрицы в квадрат
  # а далее находим среднее арифметическое элементов этой матрицы
  # важно, что нельзя ошибиться со слоями с которых берутся 2 
  # матрицы Грамма, ибо защиты от дураков нет 
  return tf.reduce_mean(tf.square(target_gram_matrix - current_style_gram_matrix))
```

```python
# общая функция вычисления потерь (J = a*jc + b*Js)
def compute_loss(model, loss_weights, init_image, gram_style_features, content_features):
  # model - сама модель, которую мы построили выше
  # loss_weights - кортеж из (style_weight, content_weight)
  # content_weight - это a из формулы выше
  # style_weight - это b из формулы выше
  # init_image - формируемое изображение
  # gram_style_features - заранее вычисленные матрицы Грамма для всех слоев
  # content_features - заранее вычисленные тензор на последнем слое VGG19
  # , описывающий исходное изображение

  style_weight, content_weight = loss_weights
 
  # пропустить формируемое изображение через нейросеть и получить выходные тензоры
  model_outputs = model(init_image)

  # так как слой для расчета ошибки по контенту - последний слой модели
  # то и в тензоре model_outputs он будет последним
  # следовательно слои для расчета ошибки по стилю и контенту можно легко
  # разделить одним срезом, зная кол-во слоев, используемых для расчета
  # ошибки по стилю
  style_output_features = model_outputs[:num_style_layers]
  content_output_features = model_outputs[num_style_layers:]

  # вспомогательные переменные для хранения величин потерь для стиля и контента
  style_score = 0
  content_score = 0

  # множитель, на который умножается каждый из слогаемых формулы
  # вычисления ошибки по стилю. Фактически усредняет сумму ошибок
  # на каждом слое
  weight_per_style_layer = 1.0 / float(num_style_layers)

  # вычислить Js (расхождение по стилю)
  # как работает zip() - https://realpython.com/python-zip-function/
  # фактически мы одновременно перебираем два списка, что удобно
  for target_style, comb_style in zip(gram_style_features, style_output_features):
      # пока что непонятно, почему comb_style[0], а не просто comb_style
      style_score += weight_per_style_layer * get_style_loss(comb_style[0], target_style)


  # множитель, на который умножается каждый из слогаемых формулы
  # вычисления ошибки по контенту. Фактически усредняет сумму ошибок
  # на каждом слое. Однако так как у нас слой 1, то это бесполезный множитель.
  weight_per_content_layer = 1.0 / float(num_content_layers)

  # вычислить Jc (расхождение по контенту)
  # цикл тоже для единообразия. Сработает только 1 раз
  for target_content, comb_content in zip(content_features, content_output_features):
    content_score += weight_per_content_layer * get_content_loss(comb_content[0], target_content)

  # вычислим общую ошибку
  content_score *= content_weight
  style_score *= style_weight

  # loss - это J из нашей формулы
  loss = content_score + style_score

  return loss, style_score, content_score

```

```python
# пропустить стилевое изображение и изображение с контентом
# через модель и получить их карты признаков
def get_feature_representation(model):
  style_outputs = model(x_style)
  content_outputs = model(x_img)

  style_features = [style_layer[0] for style_layer in style_outputs[:num_style_layers]]
  content_features = [content_layer[0] for content_layer in content_outputs[num_style_layers:]]

  return style_features, content_features
```

```python
# определение переменных перед началом обучения

num_iterations = 100
# не знаю что значит запись чисел ниже
# но это a и b из общей формулы расчета ошибки
content_weight = 1e3
style_weight = 1e-2

style_features, content_features = get_feature_representation(model)
# заранее посчитаем матрицы грамма для стилевого изображения
gram_style_features = [get_gram_matrix(style_feature) for style_feature in style_features]

# создаем начальное формируемое изображение
init_image = np.copy(x_img)
init_image = tf.Variable(init_image, dtype=tf.float32)

# создадим оптимизатор для аллгоритма градиентного спуска
opt = tf.compat.v1.train.AdamOptimizer(learning_rate=2, beta1=0.99, epsilon=1e-1)
iter_count = 1
best_loss, best_img = float('inf'), None
loss_weights = (style_weight, content_weight)

# словарь с общим описанием модели для удобства использования
cfg = {
    'model' : model,
    'loss_weights': loss_weights,
    'init_image': init_image,
    'gram_style_features': gram_style_features,
    'content_features': content_features
}

# вспомогательные переменные для преобразования изображений
# из BGR в RGB
norm_means = np.array([103.939, 116.779, 123.68])
min_vals = -norm_means
max_vals = 255 - norm_means
# колллекция, содержащая все изображения, которые мы сформируем
imgs = []
```

```python
# процесс формирования стилизованного изображения

for i in range(num_iterations):

  # ////////////////////////////////
  # духота какая то, не понимаю
  with tf.GradientTape() as tape:
    all_loss = compute_loss(**cfg)

  total_loss = all_loss[0]
  grads = tape.gradient(total_loss, init_image)

  loss, style_score, content_score = all_loss

  opt.apply_gradients([(grads, init_image)])
  clipped = tf.clip_by_value(init_image, min_vals, max_vals)
  init_image.assign(clipped)
  # ////////////////////////////////

  if loss < best_loss:
    best_loss = loss
    best_img = deprocess_img(init_image.numpy())

    plot_img = deprocess_img(init_image.numpy())
    imgs.append(plot_img)
  
  print(f'Iteration: {i}')

```
